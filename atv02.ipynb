```
# AT02 Lista - Atividade de listas em Python

# 1. Crie uma lista de compras de um supermercado com 10 elementos.
lista_compras = ["arroz", "feijão", "macarrão", "leite", "pão", "queijo", "manteiga", "café", "açúcar", "sabão"]

# 2. Na posição 6, crie uma lista com 5 elementos.
lista_compras[5] = ["alface", "tomate", "cenoura", "batata", "cebola"]

# 3. Altere 2 elementos da lista (fora da sublista)
lista_compras[0] = "arroz integral"
lista_compras[1] = "feijão preto"

# 4. Exclua 3 elementos da lista
del lista_compras[2]
del lista_compras[3]
del lista_compras[4]

# 5. Adicione 4 elementos na lista
lista_compras.append("ovo")
lista_compras.append("óleo")
lista_compras.append("sal")
lista_compras.append("farinha")

# 6. Acesse o terceiro e quinto elemento da lista
print("Terceiro elemento:", lista_compras[2])
print("Quinto elemento:", lista_compras[4])

# 7. Mostre o número de elementos da lista
print("Número de elementos da lista:", len(lista_compras))

# 8. Mostre todos os elementos da lista
print("Todos os elementos da lista:")
for item in lista_compras:
    print(item)
```
